**Hypothesis 1**

**Public schools are more than private schools, they are 3x times more than private schools.**

```
my_colors = list(islice(cycle(['orange','green']), None, len(merged_Frame)))
ax = merged_Frame.School_Type.value_counts().plot(kind='bar', figsize=(10,8),color= my_colors, fontsize=13);
ax.set_xlabel('Public or Private Schools', fontsize = '15')           # To set x axis label
ax.set_ylabel('Number of Schools', fontsize = '15')  # To set y axis label
ax.set_title('Number of Public and Private Schools', fontsize = '20')  # To set title 
ax.set_xticklabels(ax.get_xticklabels(),rotation='horizontal')
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
     width, height = i.get_width(), i.get_height()
     x, y = i.get_xy() 
     ax.annotate('{}'.format(height), (x, y + height))
```

**Conclusion of Hypothesis 1:**

**Yes, there are 3 times(3x) more Public schools than Private schools.Hence it supports my hypothesis 1 statement**

**Hypothesis 2:**

**In CO state the private schools are more when compared to other states and in the AL state the public schools are more.**

```
state_df=df.State.value_counts()

#plot the bar graph with different color to identify the highest number of public schools
plt.figure(figsize=(18,6))
for i,v in state_df.iteritems():
    if i == 'CA':
        _=plt.bar(i,v,color='red')
    else:
        _=plt.bar(i,v,color='orange')

#Setting the titles, x and y labels        
_=plt.xlabel('States',fontsize=15)
_=plt.ylabel('Number of Public Schools',fontsize=15)
_=plt.title('Number of Public Schools by different states in USA',fontsize=18)
plt.show()
```

```
state_df=df_1.State.value_counts()

#plot the bar graph with different color to identify the highest number of private schools
plt.figure(figsize=(18,6))
for i,v in state_df.iteritems():
    if i == 'CA':
        _=plt.bar(i,v,color='red')
    else:
        _=plt.bar(i,v,color='orange')

#Setting the titles, x and y labels        
_=plt.xlabel('States',fontsize=15)
_=plt.ylabel('Number of Private Schools',fontsize=15)
_=plt.title('Number of Private Schools by different states in USA',fontsize=18)
plt.show()
```

**Hypothesis 3**

**More than 70% of schools in the US are public schools.**

```
plt.figure(figsize=plt.figaspect(1))
values = [len(df),len(df_1)] 
labels = ['Public Schools','Private Schools'] 

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%'.format(p=pct)
    return my_autopct

plt.pie(values, labels=labels, autopct=make_autopct(values),colors= ['tomato','grey'])
plt.title('Percentage of Public and Private Schools in US')
plt.show()
```

**Conclusion of Hypothesis 3:**

**Yes, more than 70% Schools in US are Public Schools.Hence, it supports Hypothesis 3 statement.**

**Hypothesis 4:**

**Which state has highest number of public and private schools in US?**

```
data = (merged_Frame.groupby(['State', 'School_Type'])['School_Type'].count()).unstack('School_Type').fillna(0)
data
```

```
ax = data.plot(kind='bar', stacked = True, figsize=(15, 6), fontsize=13);
ax.set_xlabel('States', fontsize = '15')           # To set x axis label
ax.set_ylabel('No. of Schools', fontsize = '15')  # To set y axis label
ax.set_title('No.Of Public and Private Schools based on State', fontsize = '20')  # To set title 
ax.plot()
```

**Conclusion of Hypothesis 4:**

**California State has the highest number of Public and Private Schools in US, followed by Texas State.**
